# RLHF — Recorrido: Índice

Esta serie reconstruye el material de `tutorial.ipynb` como un notebook enfocado por algoritmo. Cada notebook es corto, denso en explicación, y apunta al módulo en `src/llm_rlhf/` que lo implementa.

Lee los notebooks en orden — cada uno se apoya en el anterior.

## El recorrido

| # | Notebook | Tema | Módulo |
|---|---|---|---|
| 01 | [01_pretrained.ipynb](01_pretrained.ipynb) | El modelo base y por qué no sigue instrucciones | `model.py` |
| 02 | [02_sft.ipynb](02_sft.ipynb) | Fine-tuning supervisado con LoRA sobre Dolly-15k | `sft.py` |
| 03 | [03_reward_model.ipynb](03_reward_model.ipynb) | Modelo de recompensa Bradley–Terry a partir de preferencias por pares | `reward.py` |

| 05 | [05_dpo.ipynb](05_dpo.ipynb) | Direct Preference Optimization — RLHF sin RL | `dpo.py` |
| 06 | [06_grpo.ipynb](06_grpo.ipynb) | Group Relative Policy Optimization — PPO sin función de valor | `grpo.py` |

## Convenciones en todos los notebooks

- **Un algoritmo por notebook.** Ninguno entrena un pipeline completo desde cero.
- **Primero la matemática, después el código.** Cada algoritmo se introduce con su objetivo; las celdas de código muestran la *función de pérdida*, no bucles completos de entrenamiento.
- **Visualizaciones y ejercicios.** Cada notebook incluye al menos una visualización del concepto central y un ejercicio para que pruebes hipótesis sin esperar a un entrenamiento de 15 minutos.
- **Tensores de juguete en lugar de entrenamiento real.** Cuando es posible, un tensor 2×3 demuestra la matemática. Las corridas reales son trabajo del usuario — descomenta la celda correspondiente.
- **Configs en TOML.** Cada trainer toma un dataclass; `configs/<stage>.toml` mapea sus campos vía `SFTConfig(**load_toml('configs/sft.toml'))`.
- **Solo uv.** Ejecuta `uv sync` para instalar. Lanza esta carpeta con `uv run jupyter notebook`.

## Dónde comparar algoritmos

Los módulos están diseñados para ser intercambiables: todos toman un modelo SFT y los mismos pares de preferencias `(prompt, y_w, y_l)` (PPO y GRPO necesitan además un modelo de recompensa entrenado).

Para correr una comparación lado a lado después de los notebooks 01–06, abre un notebook nuevo y replica el patrón de los bloques de código al final de cada algoritmo.

## Una nota sobre el idioma

La narrativa (markdown) está en español; el código, los nombres de variables, los prints y los identificadores técnicos se mantienen en inglés. Esto es deliberado: el código de referencia de la literatura de ML es en inglés y queremos que sea fácil cruzar entre este material y los papers originales.
